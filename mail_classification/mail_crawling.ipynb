```python
import csv
import pandas as pd
import secret
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep
```

```python
write_csv = open('val_email.csv','a',newline='')
```

```python
write = csv.writer(write_csv)
```

```python
write.writerow(['보낸 사람','제목','내용'])
```

```python
driver = webdriver.Chrome('chromedriver.exe')
```

```python
driver.get("https://nid.naver.com/nidlogin.login")
```

```python
for num in range(38900,39000):
    try: 
        driver.get(f"https://mail.naver.com/read/popup/?nMailId={num}")
        sleep(1)
        root_class01 = driver.find_element_by_class_name('viewTitle')
        title = root_class01.find_element_by_tag_name('h4').text
        sender = root_class01.find_element_by_tag_name('dd').text
        root_class02 = driver.find_element_by_id('readFrame')
        content = root_class02.text
        context = [title, sender, content]
        write.writerow(context)
    except: 
        pass
```
